```python
# ===============================
# 1. Import Libraries
# ===============================
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy.cluster.hierarchy as shc

from sklearn.preprocessing import StandardScaler
from sklearn.cluster import AgglomerativeClustering

%matplotlib inline


# ===============================
# 2. Load Dataset
# ===============================
customer_data = pd.read_csv("Country-data.csv")

print(customer_data.shape)
customer_data.head()


# ===============================
# 3. Select Features (Numerical only)
# ===============================
# Selecting: exports, health, imports
data = customer_data.iloc[:, 2:5].values


# ===============================
# 4. Feature Scaling (VERY IMPORTANT)
# ===============================
scaler = StandardScaler()
data_scaled = scaler.fit_transform(data)


# ===============================
# 5. Plot Dendrogram
# ===============================
plt.figure(figsize=(10, 7))
plt.title("Country Dendrogram")
plt.xlabel("Countries")
plt.ylabel("Euclidean Distance")

shc.dendrogram(
    shc.linkage(data_scaled, method='ward')
)

plt.show()


# ===============================
# 6. Apply Agglomerative Clustering
# ===============================
cluster = AgglomerativeClustering(
    n_clusters=6,
    linkage='ward'
)

labels_ = cluster.fit_predict(data_scaled)


# ===============================
# 7. Add Cluster Labels to DataFrame
# ===============================
customer_data['Cluster'] = labels_
customer_data.head()


# ===============================
# 8. Visualization (2D Scatter)
# ===============================
plt.figure(figsize=(10, 7))
plt.scatter(
    data_scaled[:, 0],
    data_scaled[:, 1],
    c=labels_,
    cmap='rainbow'
)

plt.xlabel("Exports (scaled)")
plt.ylabel("Health (scaled)")
plt.title("Agglomerative Hierarchical Clustering")
plt.show()


# ===============================
# 9. Cluster Interpretation
# ===============================
cluster_summary = customer_data.groupby('Cluster').mean(numeric_only=True)
print(cluster_summary)
```

# How know which country is closest to india

```python
import numpy as np
from sklearn.preprocessing import StandardScaler
from scipy.spatial.distance import euclidean
```

```python
features = customer_data.drop(columns=['country', 'Cluster'], errors='ignore')
```

```python
scaler = StandardScaler()
features_scaled = scaler.fit_transform(features)
```

```python
# Get india vector
india_index = customer_data[customer_data['country'] == 'India'].index[0]
india_vector = features_scaled[india_index]
```

```python
# compute distance to all countries
distances = []

for i in range(len(features_scaled)):
    if i != india_index:
        dist = euclidean(india_vector, features_scaled[i])
        distances.append((customer_data.loc[i, 'country'], dist))
```

```python
# find closest country
closest_country = min(distances, key=lambda x: x[1])
closest_country
```

```python
# Top 5 closest country to india
sorted(distances, key=lambda x: x[1])[:5]
```
