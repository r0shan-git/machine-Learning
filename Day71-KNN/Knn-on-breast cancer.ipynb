```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('data.csv')
df.head()
```

```python
df.shape
```

```python
df.drop(columns=['id','Unnamed: 32'],inplace=True)
df.head()
```

```python
df.shape
```

```python
# Features and label split  
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df.iloc[:,1:], df.iloc[:,0],test_size=0.2, random_state=2)
```

```python
X_train.head()
```

```python
X_train.shape
```

```python
# features scaling (agar scale nahi kiya tooh wrong distance calulate hoga isliye standard scaler mandatory)
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
X_train
```

```python
X_train.shape
```

```python
# knn model traning ( 3 neighbors use karke model train kiya
)
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors=3)
```

```python
knn.fit(X_train,y_train)
```

```python
# model evaluation(model kintna shi predict kar raha hai )
from sklearn.metrics import accuracy_score

y_pred=knn.predict(X_test)
accuracy_score(y_test,y_pred)
```

```python
#Best k value find karna(different k values ko try karke best accuracy nikalna )
#  apply cross validation 
scores=[]

for i in range(1,16):

    knn=KNeighborsClassifier(n_neighbors=i)

    knn.fit(X_train,y_train)
    y_pred=knn.predict(X_test)
    scores.append(accuracy_score(y_test,y_pred))
```

```python
import matplotlib.pyplot as plt 
plt.plot(range(1,16),scores)
plt.xlabel("K-value")
plt.ylabel("Accuracy")


# in below curve acuuracy is good at k=3
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
from sklearn import neighbors, datasets
from sklearn.preprocessing import StandardScaler
from ipywidgets import interact, fixed


# Load dataset
def load_data():
    cancer = datasets.load_breast_cancer()
    return cancer


# Plot decision boundaries
def plot_decision_boundaries(n_neighbors, data, labels):
    h = 0.02  # step size in mesh

    # Color maps
    cmap_light = ListedColormap(['orange', 'lightblue'])
    cmap_bold = ListedColormap(['darkorange', 'darkblue'])

    # KNN classifier
    clf = neighbors.KNeighborsClassifier(n_neighbors=n_neighbors)
    clf.fit(data, labels)

    # Define grid boundaries
    x_min, x_max = data[:, 0].min() - 1, data[:, 0].max() + 1
    y_min, y_max = data[:, 1].min() - 1, data[:, 1].max() + 1

    xx, yy = np.meshgrid(
        np.arange(x_min, x_max, h),
        np.arange(y_min, y_max, h)
    )

    # Predict for each point in grid
    Z = clf.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)

    # Plot decision boundary
    plt.figure(figsize=(8, 6))
    plt.pcolormesh(xx, yy, Z, cmap=cmap_light, shading='auto')

    # Plot training points
    plt.scatter(
        data[:, 0],
        data[:, 1],
        c=labels,
        cmap=cmap_bold,
        edgecolor='k',
        s=20
    )

    plt.xlabel("Feature 1 (scaled)")
    plt.ylabel("Feature 2 (scaled)")
    plt.title(f"KNN Decision Boundary (k = {n_neighbors})")
    plt.show()


# Load data
cancer = load_data()

# Use only first two features and scale them
X = StandardScaler().fit_transform(cancer.data[:, :2])
y = cancer.target

# Interactive slider
interact(
    plot_decision_boundaries,
    n_neighbors=(1, 20),
    data=fixed(X),
    labels=fixed(y)
);
```
