```python
import numpy as np 
import matplotlib.pyplot as plt 

from sklearn.linear_model import LinearRegression ,SGDRegressor

from sklearn.model_selection import train_test_split

from sklearn.preprocessing import PolynomialFeatures,StandardScaler
from sklearn.metrics import r2_score

from sklearn.pipeline import Pipeline 
```

```python
# To generate the data 
X=6 * np.random.rand(200,1)-3
y=0.8 * X**2 + 0.9*X +2 + np.random.randn(200,1)

# y=0.8x^2  + 0.9x +2
```

```python
plt.plot(X,y,'b.')
plt.ylabel("y")
plt.xlabel("X")
plt.show()
```

```python
# Train test split 
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=2)
```

```python
#Applying linear regression 
lr = LinearRegression()
```

```python
lr.fit(X_train,y_train)
```

```python
y_pred=lr.predict(X_test)
r2_score(y_test,y_pred)
```

```python
plt.plot(X_train,lr.predict(X_train),color='r')
plt.plot(X,y,"b.")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

```python
# Applying polynomial Linear Regression 
# degree 2
poly= PolynomialFeatures(degree=2,include_bias=True) # if in place true we write false then only x and x^2 term come all other are cut

X_train_trans=poly.fit_transform(X_train)
X_test_trans=poly.transform(X_test)
```

```python
print(X_train[0])
print(X_train_trans[0])
```

```python
lr=LinearRegression()
lr.fit(X_train_trans,y_train)
```

```python
y_pred=lr.predict(X_test_trans)
```

```python
r2_score(y_test,y_pred)
```

```python
print(lr.coef_)
print(lr.intercept_)
```

```python
X_new=np.linspace(-3,3,200).reshape(200,1)
X_new_poly=poly.transform(X_new)
y_new=lr.predict(X_new_poly)
```

```python
plt.plot(X_new, y_new, "r-", linewidth=2, label="Predictions")
plt.plot(X_train, y_train, "b.",label='Training points')
plt.plot(X_test, y_test, "g.",label='Testing points')
plt.xlabel("X")
plt.ylabel("y")
plt.legend()
plt.show()
```

```python
def polynomial_regression(degree):
    X_new=np.linspace(-3, 3, 100).reshape(100, 1)
    X_new_poly = poly.transform(X_new)

    polybig_features = PolynomialFeatures(degree=degree, include_bias=False)
    std_scaler = StandardScaler()  
    lin_reg = LinearRegression()
    polynomial_regression = Pipeline([    # start building a pipeline and pipelines ensures steps run in order automatically
            ("poly_features", polybig_features), # X input into polynomial features 
            ("std_scaler", std_scaler),  # scales polynomial features
            ("lin_reg", lin_reg),        # fit a linear regression model 
        ])
    polynomial_regression.fit(X, y)
    y_newbig = polynomial_regression.predict(X_new)
    plt.plot(X_new, y_newbig,'r', label="Degree " + str(degree), linewidth=2)

    plt.plot(X_train, y_train, "b.", linewidth=3)
    plt.plot(X_test, y_test, "g.", linewidth=3)
    plt.legend(loc="upper left")
    plt.xlabel("X")
    plt.ylabel("y")
    plt.axis([-3, 3, 0, 10])
    plt.show()
```

```python
polynomial_regression(5) # as ploynomial degree increases like 350 model is overfitting start 
```

```python
# Applying Gradient Descent 
poly=PolynomialFeatures(degree=2)

X_train_trans=poly.fit_transform(X_train)
X_test_trans=poly.transform(X_test)


sgd=SGDRegressor(max_iter=100)
sgd.fit(X_train_trans,y_train)


# Smooth curve for visualization
X_new = np.linspace(-2.9, 2.8, 200).reshape(200, 1)
X_new_poly = poly.transform(X_new)
y_new = sgd.predict(X_new_poly)

# Prediction on TEST set (for R²)
y_pred = sgd.predict(X_test_trans)

# Plot
plt.plot(X_new, y_new, "r-", linewidth=2,
         label="Predictions (R² = " + str(round(r2_score(y_test, y_pred), 2)) + ")")

plt.plot(X_train, y_train, "b.", label="Training points")
plt.plot(X_test, y_test, "g.", label="Testing points")

plt.legend()
plt.show()


```

```python
x = 7 * np.random.rand(100, 1) - 2.8
y = 7 * np.random.rand(100, 1) - 2.8

z = x**2 + y**2 + 0.2*x + 0.2*y + 0.1*x*y +2 + np.random.randn(100, 1)
# z = x^2 + y^2 + 0.2x + 0.2y + 0.1xy + 2
```

```python
import plotly.express as px
df = px.data.iris()  # load iris dataset into a pandas Dataframes
fig = px.scatter_3d(df, x=x.ravel(), y=y.ravel(), z=z.ravel())  # ravel mean convert multi dimensional array into 1-D
fig.show()
```

```python
lr = LinearRegression()
lr.fit(np.array([x,y]).reshape(100,2),z)

x_input = np.linspace(x.min(), x.max(), 10)
y_input = np.linspace(y.min(), y.max(), 10)
xGrid, yGrid = np.meshgrid(x_input,y_input)

final = np.vstack((xGrid.ravel().reshape(1,100),yGrid.ravel().reshape(1,100))).T

z_final = lr.predict(final).reshape(10,10)
```

```python
import plotly.graph_objects as go

fig = px.scatter_3d(df, x=x.ravel(), y=y.ravel(), z=z.ravel())

fig.add_trace(go.Surface(x = x_input, y = y_input, z =z_final ))

fig.show()
```
