```python
from sklearn.datasets import make_regression
import pandas as pd
import numpy as np

import plotly.express as px
import plotly.graph_objects as go

from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
```

```python
X,y = make_regression(n_samples=100, n_features=2, n_informative=2, n_targets=1, noise=50)  # make regression :-- meaning how many data you want to create 
```

```python
df = pd.DataFrame({'feature1':X[:,0],'feature2':X[:,1],'target':y})
```

```python
df.shape
```

```python
df.head()
```

```python
fig = px.scatter_3d(df, x='feature1', y='feature2', z='target')

fig.show()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=3)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
lr.fit(X_train,y_train)
```

```python
y_pred = lr.predict(X_test)
```

```python
print("MAE",mean_absolute_error(y_test,y_pred))
print("MSE",mean_squared_error(y_test,y_pred))
print("R2 score",r2_score(y_test,y_pred))
```

```python
import numpy as np

# 1 Create values
x = np.linspace(-5, 5, 10)  # create  10 equally spaced values
y = np.linspace(-5, 5, 10)

# 2 Create grid
xGrid, yGrid = np.meshgrid(x, y) # create 2D plot grid od points

# 3 Create input matrix for model (IMPORTANT)
final = np.vstack((
    xGrid.ravel(),   #  Converts 2D grid  1D array (100 points)
    yGrid.ravel()
)).T    # shape = (100, 2)

# 4 Predict using trained model
z_final = lr.predict(final).reshape(10, 10)

# 5 Assign (optional)
z = z_final
```

```python
fig = px.scatter_3d(df, x='feature1', y='feature2', z='target')

fig.add_trace(go.Surface(x = x, y = y, z =z ))

fig.show()
```
